```python
import pandas as pd
import os
import json


df = pd.read_excel(R"D:\testcode  python\qipashuo\时间戳_完整版.xlsx")
df.head()
```

```python
st = set(df[df["季度"] == 2]["排期"])
st
```

```python
range_dict = {episode: {key: [val] for key, val in zip(df[df["排期"] == episode]["timestamps"], df[df["排期"] == episode]["辩手姓名"])} for episode in st}
range_dict
```

```python
for root, dirs, files in os.walk(R"D:\testcode  python\qipashuo\bullet_season2"):
    for file in files:
        json_file = os.path.join(root, file)
        if json_file.endswith(".json"):
            with open(json_file, 'r', encoding='utf-8') as f:
                json_data = json.load(f)
                idx = 0
                range_keys = list(range_dict[f"2-{file[:2]}-0"].keys())
                left, right = map(int, range_keys[idx].split("-"))
                for tp in sorted(json_data, key=lambda x: x[1]):
                    if int(tp[1]) < left:
                        continue
                    if left <= int(tp[1]) <= right:
                        range_dict[f"2-{file[:2]}-0"][f"{left}-{right}"].append(tp[0])
                    else:
                        idx += 1
                        if idx == len(range_keys):
                            break
                        left, right = map(int, range_keys[idx].split("-"))
                        range_dict[f"2-{file[:2]}-0"][f"{left}-{right}"].append(tp[0])
```

```python
for k, dic in range_dict.items():
    for key, ls in dic.items():
        pos, neg = 0, 0
        for word in ls[1:]:
            if word == "Negative":
                neg += 1
            elif word == "Positive":
                pos += 1
        if pos + neg != 0:
            range_dict[k][key] = [ls[0], f"Posituve:{pos}", f"Negative:{neg}", (pos - neg)/(pos + neg)]
range_dict
```

```python
# 创建一个空列表用于存储结果行
rows = []

# 遍历第一层键
for first_key, second_dict in range_dict.items():
    # 遍历第二层键
    for second_key, values in second_dict.items():
        if len(values) == 1:  # 如果只有名字，添加默认值
            name = values[0]
            pos_count = ''
            neg_count = ''
            score = ''
        else:
            name = values[0]
            pos_str = values[1]
            neg_str = values[2]
            score = str(values[-1])
            
            pos_count = pos_str.split(':')[1]
            neg_count = neg_str.split(':')[1]

        rows.append([first_key, second_key, name, pos_count, neg_count, score])

# 创建 DataFrame
df = pd.DataFrame(rows, columns=['排期', '时间戳', '姓名', 'Pos个数', 'Neg个数', '计算值'])

# 将 DataFrame 写入 CSV 文件
df.to_csv(R'D:\testcode  python\qipashuo\弹幕.csv', index=False, encoding='utf-8')

print("CSV 文件已成功生成")
```
